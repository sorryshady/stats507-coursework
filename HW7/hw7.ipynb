## Question 1.8 (4 pts)
Segment the dep_time 24 hours into 
* night from 12 midnight to 5 am
* morning from 5:01 am to 12 noon
* afternoon from 12.01 noon to 5 pm
* evening from 5.01 pm to 12 am

And then run a query to find the total flights in each segment and save that value in the given variable.


````python
# Segment dep_time into parts of day
def segment_time(dep_time):
    if pd.isna(dep_time):
        return None
    # Convert to integer for comparison
    dep_time_int = int(dep_time)
    hour = dep_time_int // 100
    minute = dep_time_int % 100
    
    # night: 12 midnight to 5 am (00:00-05:00)
    if hour >= 0 and hour < 5:
        return 'night'
    elif hour == 5 and minute == 0:
        return 'night'
    # morning: 5:01 am to 12 noon (05:01-12:00)
    elif hour == 5 and minute > 0:
        return 'morning'
    elif hour > 5 and hour < 12:
        return 'morning'
    elif hour == 12 and minute == 0:
        return 'morning'
    # afternoon: 12:01 noon to 5 pm (12:01-17:00)
    elif hour == 12 and minute > 0:
        return 'afternoon'
    elif hour > 12 and hour < 17:
        return 'afternoon'
    elif hour == 17 and minute == 0:
        return 'afternoon'
    # evening: 5:01 pm to 12 am (17:01-23:59)
    else:
        return 'evening'

df['time_segment'] = df['dep_time'].apply(segment_time)
parts_of_day_count = df['time_segment'].value_counts().to_dict()
````

````python
assert parts_of_day_count['morning'] == 129939
assert parts_of_day_count['evening'] == 98362
````

## Question 2.3 (3 pts)

Segment the flights based on time-slots with the following criteria
* flights that depart from 00-5:59am as 'red-eye' flights
* flights that depart from 6am - 11:59am as 'morning'
* flights that depart from 12noon - 3:59pm as 'afternoon'
* flights that depart from 4pm - 11:59pm as 'evening' 

And using seaborn create a plot as shown below. PS: Feel free to choose your color palette.

segmentation.png

````python
# Helper function for time range labels
def get_time_range(slot):
    ranges = {
        'red-eye': '00-5:59am',
        'morning': '6am-11:59am',
        'afternoon': '12noon-3:59pm',
        'evening': '4pm-11:59pm'
    }
    return ranges.get(slot, '')

# Segment flights based on time-slots
def segment_time_slot(dep_time):
    if pd.isna(dep_time):
        return None
    dep_time_int = int(dep_time)
    hour = dep_time_int // 100
    
    if hour >= 0 and hour < 6:
        return 'red-eye'
    elif hour >= 6 and hour < 12:
        return 'morning'
    elif hour >= 12 and hour < 16:
        return 'afternoon'
    else:  # hour >= 16 and hour < 24
        return 'evening'

df['time_slot'] = df['dep_time'].apply(segment_time_slot)

# Create seaborn plot with subplots for each time slot showing flights by origin
import seaborn as sns

time_slots = ['red-eye', 'morning', 'afternoon', 'evening']
fig_3 = sns.catplot(data=df[df['time_slot'].notna()], 
                    x='origin', 
                    kind='count',
                    hue='origin',
                    col='time_slot',
                    col_wrap=4,
                    col_order=time_slots,
                    palette='Set2')

# Set titles and labels
for i, ax in enumerate(fig_3.axes.flat):
    time_slot = time_slots[i]
    ax.set_title(f'{time_slot}({get_time_range(time_slot)})')
    ax.set_ylabel('Flight Count')
    ax.set_xlabel('')

plt.tight_layout()
plt.show()
````

````python
assert fig_3.figure.get_axes()[0].get_title() == 'red-eye(00-5:59am)'
assert fig_3.figure.get_axes()[0].get_ylabel() == 'Flight Count'
assert fig_3.figure.get_axes()[1].get_title() == 'morning(6am-11:59am)'
````
